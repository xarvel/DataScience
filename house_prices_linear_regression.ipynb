```python
features_columns = []

features_columns.extend(MSSubClass_columns)
features_columns.extend(MSZoning_columns)
features_columns.append('LotFrontage');
features_columns.append('LotArea');
features_columns.extend(Foundation_columns)
features_columns.extend(Heating_columns)
features_columns.append('GarageCars');
features_columns.append('HasPool');
features_columns.append('GrLivArea');
features_columns.extend(Neighborhood_columns)
features_columns.append('CentralAir');
features_columns.append('YearBuilt');
features_columns.append('OverallCond');
features_columns.append('OverallQual');

#features_columns.extend(RoofMatl_columns)
features_columns.extend(RoofStyle_columns)
features_columns.extend(Electrical_columns)
```

```python
from sklearn.svm import SVC
from sklearn import linear_model

regr = linear_model.LinearRegression()

from sklearn.model_selection import train_test_split
train, test = train_test_split(data_normalized, test_size=0.2, random_state=42)

train_Y = train[['SalePrice']]
train_X = train[features_columns]

test_Y = test[['SalePrice']]
test_X = test[features_columns]

regr.fit(train_X, train_Y)
regr.score(test_X, test_Y)
```

```python
regr_preditions = regr.predict(test_X).flatten() * source["SalePrice"].max()
test_Y_real = test_Y.values.flatten() * source["SalePrice"].max()

plt.scatter(test_Y_real, regr_preditions)
plt.plot([0, get_price(1)], [0, get_price(1)])
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=10) 
neigh.fit(train_X, train_Y)
neigh.score(test_X, test_Y)
```

```python
from sklearn import svm

regr = svm.SVR()
regr.fit(train_X, train_Y)
regr.score(test_X, test_Y)
```

```python
import tensorflow as tf
import keras.backend as K

def create_model():
  return tf.keras.Sequential([
        tf.keras.layers.Dense(100, input_dim=train_X.shape[1], activation='tanh',kernel_initializer='normal'),
        tf.keras.layers.Dense(50, activation='relu',kernel_initializer='normal'),
        tf.keras.layers.Dense(20, activation='relu',kernel_initializer='normal'),
        tf.keras.layers.Dense(1, activation='linear')
  ])
  
optimizer = tf.keras.optimizers.RMSprop(0.01)  
model = create_model()

def soft_acc(y_true, y_pred):
    return K.mean(K.equal(K.round(y_true), K.round(y_pred)))

def coeff_determination(y_true, y_pred):
    SS_res =  K.sum(K.square( y_true-y_pred )) 
    SS_tot = K.sum(K.square( y_true - K.mean(y_true) ) ) 
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )

model.compile(optimizer=optimizer,
              loss='mse', 
              metrics=[soft_acc, coeff_determination])
  
model.summary()
```

```python
history = model.fit(
            train_X, 
            train_Y,
            #validation_split = 0.2,
            validation_data=(test_X, test_Y),
            batch_size=50,
            epochs=100)
```

```python
results = model.evaluate(test_X, test_Y, batch_size=100)
print(results)
```
