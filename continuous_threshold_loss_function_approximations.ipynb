```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def threshold(M):
    return np.array(M < 0, dtype=np.int)

def V(M):
    return np.maximum(0, 1 - M)

def H(M):
    return np.maximum(0, -M)

def L(M):
    return np.log2(1 + np.exp(-M))

def Q(M):
    return np.power((1 - M), 2)

def E(M):
    return np.exp(-M)

def S(M):
    return 2 * np.power((1 + np.exp(M)), -1)    

M = np.arange(-5.0, 5.0,0.1)
y_threshold = threshold(M)
y_V = V(M)
y_H = H(M)
y_Q = Q(M)
y_E = E(M)
y_S = S(M);
y_L = L(M);

plt.plot(M, y_threshold,  color='red')
plt.plot(M, y_V, label='V', color='midnightblue')
plt.plot(M, y_H, label='H', color='dodgerblue')
plt.plot(M, y_L, label='L', color='violet')
plt.plot(M, y_Q, label='Q', color='lime')
plt.plot(M, y_E, label='E', color='gold')
plt.plot(M, y_S, label='S', color='darkgreen')

plt.axhline(y=0, color='k',linewidth=1)
plt.axvline(x=0, color='k',linewidth=1)

plt.xlim(-5, 5)
plt.ylim(0, 5)
plt.legend()
plt.show()
```

$V(M) = (1 - M)_+$ - кусочно-линейная (SVM);

$H(M) = (-M)_+$ - кусочно-линейная (Hebb's rule);

$L(M) = log_2(1 - e^{-M})$ - логарифмическая (LR);

$Q(M) = (1 - M)^2$ - квадратичная (FR);

$S(M) = 2(1 + e^M)^{-1}$ - сигмоидная (ANN);

$E(M) = e^{-M}$ - экспоненциальная (AdaBoost);

$[M < 0]$ - пороговая функция потерь;
