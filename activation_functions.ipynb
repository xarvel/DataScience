```
import numpy as np
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
```

```
def step(x):
    return np.array(x > 0, dtype=np.int)

def sigmoid(x): #logistic
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)
  
def softmax(w, t = 0.5):
    e = np.exp(np.array(w) / t)
    dist = e / np.sum(e)
    return dist
  
def logistic(x):
  return 1 / (1 + np.exp(-x))

def linear(x):
  return x

def tanh(x):
  return np.tanh(x)

def safe_softplus(x):
  return np.log(1.0 + np.exp(x))
    

x = np.arange(-5.0, 5.0,0.1)
y_step = step(x)
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_softmax = softmax(x)
y_linear = linear(x)
y_tanh = tanh(x)
y_softplus = safe_softplus(x)



plt.plot(x, y_step, label='Step', color='red')
plt.plot(x, y_sigmoid, label='Sigmoid', color='green')
plt.plot(x, y_relu, label='ReLU', color='blue')
plt.plot(x, y_softmax, label='softmax', color='yellow')
#plt.plot(x, y_linear, label='Linear or Identity', color='cyan')
plt.plot(x, y_tanh, label='Tanh', color='magenta')
plt.plot(x, y_softplus, label='softplus', color='violet')

plt.axhline(y=0, color='k',linewidth=1)
plt.axvline(x=0, color='k',linewidth=1)

plt.ylim(-1.1, 1.1)
plt.legend()
plt.show()
```
