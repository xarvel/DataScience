```
import numpy as np
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
```

```
def step(x):
    return np.array(x > 0, dtype=np.int)

def sigmoid(x): #logistic
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)
  
def softmax(w, t = 0.5):
    e = np.exp(np.array(w) / t)
    dist = e / np.sum(e)
    return dist
  
def logistic(x):
  return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0,0.1)
y_step = step(x)
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_softmax = softmax(x)

plt.plot(x, y_step, label='Step', color='red')
plt.plot(x, y_sigmoid, label='Sigmoid', color='green')
plt.plot(x, y_relu, label='ReLU', color='blue')
plt.plot(x, y_softmax, label='softmax', color='yellow')

plt.ylim(-0.1, 1.1)
plt.legend()
plt.show()
```
